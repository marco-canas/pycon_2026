### Semana 1: El "Bug" de Origen — ¿Por qué los LLMs son vulnerables por diseño?

**Objetivo:** Comprender por qué los LLMs no se pueden parchar como el software tradicional y qué significa que sean "procesos estocásticos".

* **Conceptos clave:**
* **Estocasticidad y Autorregresión:** Cómo predice un LLM el siguiente token y por qué la aleatoriedad controlada (*temperature*) impide el determinismo.
* **El lenguaje como vector de ataque:** Por qué las instrucciones de control (system prompts) y los datos de usuario comparten el mismo canal de entrada (contexto), haciendo imposible una separación dura de privilegios.


* **Práctica:**
1. Lee el paper clásico de Andrej Karpathy sobre las limitaciones de los LLMs o mira su video *Intro to Large Language Models*.
2. Experimenta en un Playground (OpenAI/Anthropic) intentando que un modelo ignore sus instrucciones iniciales usando técnicas de ingeniería de prompts sencillas.



---

### Semana 2: El Arsenal del Atacante y los Estándares de la Industria

**Objetivo:** Conocer los tipos de ataques del mundo real y los marcos de trabajo que Andrés mencionará (OWASP y MLCommons).

* **Conceptos clave:**
* **Inyección de Prompts (Prompt Injection):** Directa (el usuario engaña al chatbot) e Indirecta (el LLM lee un documento web malicioso y ejecuta órdenes ocultas).
* **Fugas de Datos (Data Leakage / Exfiltration):** Conseguir que el modelo revele su prompt de sistema o datos sensibles del RAG.
* **Estándares de Evaluación:**
* **OWASP LLM Top 10:** El estándar de oro de vulnerabilidades en aplicaciones LLM.
* **MLCommons AI Safety:** Cómo se mide el comportamiento seguro de un modelo bajo estrés.




* **Práctica:**
1. Estudia a fondo el documento oficial de **OWASP Top 10 for LLM Applications**. Enfoque especial en: *LLM01: Prompt Injection* y *LLM06: Sensitive Information Disclosure*.
2. Explora el caso real de **Air Canada** (mencionado en la charla) para entender cómo el chatbot inventó una política de reembolsos y la empresa fue declarada legalmente responsable.



---

### Semana 3: Rompiendo Modelos a Escala (Adversarial Testing)

**Objetivo:** Entender la metodología detrás de benchmarks como el *llm-break-bench* que Andrés presentará.

* **Conceptos clave:**
* **Jailbreaking:** Técnicas complejas para saltarse los filtros de alineación del modelo (juegos de rol, codificación en Base64, ataques multilingües).
* **Red Teaming Automatizado:** Uso de algoritmos o de otros LLMs para generar miles de ataques sistemáticos a un modelo objetivo.
* **La paradoja del costo/tamaño:** Por qué los modelos más grandes y costosos no siempre son los más seguros (a mayor razonamiento, a veces es más fácil convencerlos con argumentos lógicos retorcidos).


* **Práctica:**
1. Lee sobre **Adversarial Robustness** en IA.
2. Intenta diseñar un script simple en Python que envíe 10 variaciones de un prompt prohibido a una API de LLM para ver en qué porcentaje logras evadir su filtro nativo.



---

### Semana 4: Defensas de Diseño (Guardrails & Firewalls)

**Objetivo:** Aprender a proteger tus aplicaciones sin depender del "buen comportamiento" del modelo. Esta es la parte más accionable de la charla.

* **Conceptos clave:**
* **NVIDIA NeMo Guardrails:** Cómo crear rieles de seguridad basados en flujos (*Colang*) para interceptar entradas de usuario y salidas de modelo antes de que causen daño.
* **Meta Llama Guard / LlamaFirewall:** Modelos clasificadores dedicados exclusivamente a detectar contenido tóxico o instrucciones maliciosas en milisegundos.
* **La regla de oro del Diseño Responsable:** Tratar al LLM como una "caja negra insegura". Toda validación crítica debe ocurrir *fuera* del modelo.


* **Práctica:**
1. Sigue el tutorial de inicio rápido de **NVIDIA NeMo Guardrails** en GitHub. Construye un bot de RAG sencillo que se niegue a hablar de política o competidores.
2. Investiga cómo integrar un modelo de moderación ligero (como *Llama Guard*) como middleware en una API de Python (FastAPI).
