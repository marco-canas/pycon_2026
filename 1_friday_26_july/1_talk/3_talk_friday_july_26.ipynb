# 📘 Fase 1: Fundamentos de LLMs y Consumo de Tokens

Para optimizar costos, primero debemos entender la "moneda de cambio" de la Inteligencia Artificial: **el token**. Los modelos de lenguaje no leen palabras como los humanos, ni bytes como las computadoras tradicionales.



## 1. El Concepto Clave: ¿Qué es un Token?

Un token es la unidad mínima de procesamiento que utiliza un LLM. Puedes verlo como un fragmento de palabra.

> 💡 **Regla general para el idioma inglés:**
> `1 Token ≈ 4 caracteres` o `0.75 palabras`.
> Por lo tanto, 100 palabras son aproximadamente 133 tokens.



# El "Impuesto del Español" (The Spanish Tax)

Aquí está el primer gran secreto que un arquitecto de software debe saber: **El español es más caro que el inglés en IA.**
Los tokenizadores (los algoritmos que parten el texto en tokens) fueron entrenados mayoritariamente con texto en inglés. Al procesar inglés, suelen asignar un token por palabra completa. En español, al no reconocer las raíces de las palabras o los acentos de la misma forma, fragmentan las palabras en múltiples pedazos.

* *Inglés:* `Tomorrow` $\rightarrow$ **1 token**
* *Español:* `Mañana` $\rightarrow$ Puede costar **2 o 3 tokens** dependiendo del modelo.

Si tu aplicación atiende a usuarios en Latinoamérica o España, tu factura de tokens puede ser entre un **30% y un 50% más alta** que la de una app idéntica en Estados Unidos solo por el idioma.

---



## 2. Anatomía del Costo: Input vs. Output

Cuando consumes un LLM a través de una API (como Amazon Bedrock o OpenAI), la tarifa se divide en dos conceptos con precios radicalmente distintos:

1. **Tokens de Entrada (Prompt/Input Tokens):** Todo lo que tú le envías al modelo (Instrucciones del sistema + Historial de la conversación + Contexto extraído de documentos + La pregunta actual del usuario). **Es más barato.**
2. **Tokens de Salida (Completion/Output Tokens):** La respuesta que el modelo genera palabra por palabra. **Es entre 3 y 4 veces más costosa** porque requiere que el modelo gaste poder de cómputo en tiempo real para predecir cada siguiente fragmento.

$$Costo_{Total} = (Tokens_{In} \times Tarifa_{In}) + (Tokens_{Out} \times Tarifa_{Out})$$



### La Ventana de Contexto (Context Window)

Es el límite máximo de tokens que el modelo puede "recordar" en una sola sesión. Si tu ventana es de 128,000 tokens, puedes meter un libro entero dentro. Pero cuidado: **en la arquitectura tradicional, pagas por cada token que metes en cada interacción.** Si el usuario hace una pregunta de 5 palabras, pero tú le inyectas un historial de 10,000 tokens atrás para que "tenga contexto", vas a pagar esos 10,000 tokens *en cada mensaje nuevo*. Aquí es donde el presupuesto se desangra.

## 3. Práctica en Python: Contar Tokens y Estimar Costos

Para no depender de adivinanzas, vamos a usar Python para calcular los tokens exactos antes de enviarlos a la nube. Usaremos la librería `tiktoken` (creada por OpenAI, pero que sirve como excelente base para entender el concepto).

### Paso 2: El Script de Python (`cost_estimator.py`)

Crea un archivo con el siguiente código. Este script simula el procesamiento de un texto y calcula cuánto te costaría en producción usando tarifas estándar de la industria (por ejemplo, estimando tarifas hipotéticas promedio de $0.0015 por cada 1,000 tokens de entrada y $0.0045 por cada 1,000 de salida).



````python
import tiktoken

def analizar_costo_prompt(texto_prompt: str, texto_respuesta_estimada: str):
    # 1. Seleccionar el tokenizador del modelo (ej. cl100k_base usado por GPT-3.5/4)
    # Modelos de AWS Bedrock como Claude usan sus propios tokenizadores (ej. anthropic), 
    # pero tiktoken nos da la intuición matemática perfecta.
    tokenizador = tiktoken.get_encoding("cl100k_base")
    
    # 2. Convertir texto a tokens (Encoding)
    tokens_input = tokenizador.encode(texto_prompt)
    tokens_output = tokenizador.encode(texto_respuesta_estimada)
    
    num_tokens_in = len(tokens_input)
    num_tokens_out = len(tokens_output)
    
    # 3. Definir tarifas por cada 1,000,000 de tokens (Precios de referencia)
    PRECIO_INPUT_POR_MILLON = 1.50   # USD
    PRECIO_OUTPUT_POR_MILLON = 4.50  # USD
    
    # 4. Calcular costos
    costo_input = (num_tokens_in / 1_000_000) * PRECIO_INPUT_POR_MILLON
    costo_output = (num_tokens_out / 1_000_000) * PRECIO_OUTPUT_POR_MILLON
    costo_total = costo_input + costo_output
    
    # 5. Mostrar resultados
    print("--- REPORTE DE TOKENS Y COSTOS ---")
    print(f"Texto de entrada: '{texto_prompt[:40]}...' ({len(texto_prompt)} caracteres)")
    print(f"🔹 Tokens de Entrada: {num_tokens_in}")
    print(f"🔹 Tokens de Salida: {num_tokens_out}")
    print(f"Total de Tokens del ciclo: {num_tokens_in + num_tokens_out}")
    print("-" * 34)
    print(f"💰 Costo Entrada: ${costo_input:.6f} USD")
    print(f"💰 Costo Salida:  ${costo_output:.6f} USD")
    print(f"🚀 Costo Total:   ${costo_total:.6f} USD")
    
    # Mostrar cómo el modelo ve el texto realmente (IDs numéricos)
    print(f"\nVisualización de los primeros 5 tokens de entrada: {tokens_input[:5]}")

# --- PRUEBA EL SCRIPT ---
prompt_usuario = "Hola, necesito que analices este reporte financiero detallado de la empresa para el trimestre actual y me digas si vamos por buen camino."
respuesta_modelo = "Basado en los datos provistos, la empresa muestra un crecimiento saludable del 12% en ingresos operativos respecto al trimestre anterior."

analizar_costo_prompt(prompt_usuario, respuesta_modelo)

````

# 📘 Fase 2: Python Avanzado (Arbitraje de Modelos y Asincronismo)

En una arquitectura tradicional (y costosa), todas las preguntas de los usuarios se envían al modelo más potente y caro disponible (por ejemplo, Claude 3.5 Sonnet o GPT-4o).

En una arquitectura eficiente, tu código Python actúa como un **enrutador inteligente**. Si un usuario solo dice *"Hola, buenas tardes"*, no necesitas gastar poder de procesamiento de grado militar para responderle; usas un modelo miniatura, ultra rápido y casi gratuito. Solo escalas la petición al "modelo inteligente" si la tarea lo amerita.

---

## 1. El Concepto Clave: Arbitraje de Modelos

El arbitraje consiste en evaluar la complejidad de una tarea antes de procesarla. Para implementarlo sin agregar latencia extrema, se suele usar un **LLM Clasificador** muy pequeño o reglas heurísticas.

### La Matriz de Decisión Operativa

Para diseñar esta lógica, los arquitectos dividen los modelos en "Tiers" o niveles:

| Nivel (Tier) | Capacidad | Costo Promedio (por millón de tokens) | Caso de Uso Ideal | Ejemplos (AWS Bedrock) |
| --- | --- | --- | --- | --- |
| **Tier 1: Ligero** | Velocidad pura, tareas sencillas | ~ $0.15 - $0.30 USD | Saludos, clasificar intenciones, formatear JSON, extraer entidades. | Amazon Titan Text Lite, Claude 3 Haiku, Llama 3.1 8B |
| **Tier 2: Avanzado** | Razonamiento complejo, matemáticas, código | ~ $3.00 - $15.00 USD | Análisis financiero, generación de código, resúmenes extensos de documentos. | Claude 3.5 Sonnet, Llama 3.1 405B |

---

## 2. Flujos Asíncronos (`asyncio`)

Cuando haces una petición a un LLM en la nube, tu programa se queda "esperando" la respuesta de la red durante 2, 5 o hasta 10 segundos. Si usas código síncrono tradicional, tu servidor se congela y no puede atender a ningún otro usuario mientras tanto.

Para que una aplicación de IA sea **escalable**, todo el pipeline de Python debe ser asíncrono (`async`/`await`). Esto permite que un solo servidor procese cientos de peticiones de IA al mismo tiempo optimizando los recursos de infraestructura al máximo.

---

## 3. Práctica en Python: El Enrutador Inteligente Asíncrono

Vamos a simular el núcleo de un sistema de arbitraje de modelos utilizando programación asíncrona. Como aún no estamos conectados a AWS Bedrock (eso es la Fase 4), simularemos las llamadas a las APIs usando funciones asíncronas con tiempos de espera reales (`asyncio.sleep`).

### El Script de Python (`model_arbitrage.py`)

Crea un nuevo archivo en tu carpeta llamado `model_arbitrage.py` y pega el siguiente código:

```python
import asyncio
import time

# SIMULACIÓN DE COSTOS (Precios por 1M de tokens)
COSTO_HAIKU_LIGERO = 0.25   # Económico
COSTO_SONNET_POTENTE = 3.00  # 12 veces más caro

async def clasificar_complejidad_tarea(input_usuario: str) -> str:
    """
    Simula un modelo clasificador rápido (Tier 1) que evalúa la pregunta.
    Retorna 'SIMPLE' o 'COMPLEJA'.
    """
    await asyncio.sleep(0.2)  # Simula latencia ultra rápida de un modelo pequeño
    
    # Heurística simple para la simulación: palabras clave que denotan complejidad
    palabras_complejas = ["analiza", "calcula", "financiero", "codigo", "programar", "optimiza"]
    
    if any(palabra in input_usuario.lower() for palabra in palabras_complejas):
        return "COMPLEJA"
    return "SIMPLE"

async def ejecutar_modelo_ligero(prompt: str):
    """Simula una llamada a un modelo Tier 1 (Barato y rápido)"""
    await asyncio.sleep(0.5)  # Latencia baja
    return "Contestando con Modelo Ligero (Claude Haiku): ¡Hola! Claro que sí, dime en qué te ayudo."

async def ejecutar_modelo_potente(prompt: str):
    """Simula una llamada a un modelo Tier 2 (Costoso y potente)"""
    await asyncio.sleep(2.0)  # Los modelos grandes tardan más en razonar
    return "Contestando con Modelo Potente (Claude Sonnet): Procediendo al análisis matemático profundo de las variables financieras..."

async def orquestador_inteligente(id_usuario: int, prompt: str):
    """
    Este es el núcleo de la charla de Juan Diego:
    Decidir dinámicamente el motor de inteligencia.
    """
    inicio_peticion = time.time()
    
    # 1. Evaluamos la complejidad de manera asíncrona
    complejidad = await clasificar_complejidad_tarea(prompt)
    
    # 2. Arbitraje de modelos (Decisión financiera)
    if complejidad == "SIMPLE":
        modelo_usado = "Tier 1 (Económico)"
        costo_estimado = COSTO_HAIKU_LIGERO
        respuesta = await ejecutar_modelo_ligero(prompt)
    else:
        modelo_usado = "Tier 2 (Premium)"
        costo_estimado = COSTO_SONNET_POTENTE
        respuesta = await ejecutar_modelo_potente(prompt)
        
    tiempo_total = time.time() - inicio_peticion
    
    print(f"[Usuario {id_usuario}] 🔀 Ruta elegida: {modelo_usado} | ⏱️ Tiempo: {tiempo_total:.2f}s | 💰 Multiplicador de Costo base: {costo_estimado} USD/M")
    print(f"[Usuario {id_usuario}] 🤖 Respuesta: {respuesta}\n")

async def main():
    print("🚀 Iniciando servidor de IA con Arbitraje de Modelos...\n")
    inicio_servidor = time.time()
    
    # Simulamos que entran 3 peticiones de usuarios al mismo tiempo
    peticiones = [
        orquestador_inteligente(1, "Hola, buenas tardes, ¿cómo estás?"),
        orquestador_inteligente(2, "Analiza este reporte trimestral y calcula el ROI optimizado"),
        orquestador_inteligente(3, "Dame un saludo de bienvenida corto")
    ]
    
    # Ejecutamos las 3 peticiones EN PARALELO gracias a asyncio
    await asyncio.gather(*peticiones)
    
    print(f"⏱️ El servidor procesó TODAS las peticiones concurrentes en: {time.time() - inicio_servidor:.2f} segundos.")

# Ejecutar el bucle asíncrono
if __name__ == "__main__":
    asyncio.run(main())

```

# 📘 Fase 3: Bases de Datos Vectoriales y Caché Semántico (Reutilizar en lugar de Gastar)

El almacenamiento en caché tradicional (como el que usan las páginas web) funciona por coincidencia exacta. Si guardas la respuesta para *"¿Cómo cambio mi contraseña?"*, y otro usuario pregunta *"¿Cómo cambio mi contraseña?"*, el caché funciona. Pero si pregunta *"¿De qué forma modifico la clave?"*, el caché tradicional falla porque las palabras cambiaron, aunque el **significado** sea exactamente el mismo.

El **Caché Semántico** resuelve esto usando matemáticas e Inteligencia Artificial para entender el significado subyacente.

---

## 1. El Concepto Clave: Embeddings y Distancia Coseno

Para que una computadora entienda el "significado", usamos un modelo de **Embeddings**. Este modelo toma un texto y lo convierte en una lista de números flotantes (un vector matemático).

* Si dos frases significan lo mismo, sus vectores estarán muy "cerca" en el espacio geométrico.
* Si significan cosas distintas, sus vectores estarán "lejos".

La cercanía se mide con un algoritmo matemático llamado **Similitud de Coseno** u otras métricas de distancia (como la distancia Euclidiana).

> 🎯 **El Pipeline del Caché Semántico:**
> 1. El usuario hace una pregunta.
> 2. Convertimos la pregunta en un vector (Embedding).
> 3. Buscamos en una **Base de Datos Vectorial** local si existe alguna pregunta guardada que sea un **90% o más similar**.
> 4. **¿Existe?** Devolvemos la respuesta guardada al instante. Costo: **$0 USD**, Latencia: **Milisegundos**.
> 5. **¿No existe?** Llamamos al LLM, guardamos su respuesta en la base de datos vectorial para el futuro, y se la entregamos al usuario.

## 2. Práctica en Python: Construyendo un Caché Semántico desde Cero

Para mantener esto liviano y entender la matemática detrás sin meter herramientas pesadas aún, utilizaremos un enfoque nativo con Python. Simularemos la generación de embeddings y utilizaremos la lógica vectorial para activar o evadir el costo del LLM.

### El Script de Python (`semantic_cache.py`)

Crea un archivo llamado `semantic_cache.py` y pega el siguiente código:

```python
import math
import time

# SIMULACIÓN DE EMBEDDINGS (Vectores de significado simplificados a 3 dimensiones)
# En la vida real, un modelo de embeddings (como los de AWS Bedrock) genera vectores de 1536 dimensiones.
DICCIONARIO_EMBEDDINGS = {
    "¿cómo restauro mi clave de acceso?": [0.91, 0.12, 0.05],
    "¿de qué forma puedo cambiar la contraseña?": [0.89, 0.14, 0.04], # Muy similar a la anterior
    "¿cuánto cuesta la suscripción mensual?": [0.02, 0.88, 0.45],      # Completamente diferente
    "¿cuál es el precio del plan por mes?": [0.03, 0.85, 0.48]        # Muy similar a la de costos
}

# Base de datos vectorial en memoria (Simulando FAISS o ChromaDB)
# Estructura: {"pregunta_original": {"vector": [...], "respuesta_llm": "..."}}
BASE_DATOS_VECTORIAL_CACHE = {
    "¿cómo restauro mi clave de acceso?": {
        "vector": [0.91, 0.12, 0.05],
        "respuesta_llm": "Para restaurar tu clave, ve a la pantalla de inicio, haz clic en 'Olvidé mi contraseña' y sigue el enlace enviado a tu correo corporativo."
    }
}

def calcular_similitud_coseno(vec1, vec2):
    """Calcula qué tan similares son dos vectores (Resultado entre 0.0 y 1.0)"""
    dot_product = sum(a * b for a, b in zip(vec1, vec2))
    norm_a = math.sqrt(sum(a * a for a in vec1))
    norm_b = math.sqrt(sum(b * b for b in vec2))
    if norm_a == 0 or norm_b == 0:
        return 0.0
    return dot_product / (norm_a * norm_b)

def buscar_en_cache_semantico(nueva_pregunta: str, umbral_tolerancia=0.95):
    """
    Busca si la nueva pregunta tiene un significado similar 
    a algo que ya respondimos en el pasado.
    """
    # 1. Obtener el embedding de la nueva pregunta (Simulado de nuestra base)
    nuevo_vector = DICCIONARIO_EMBEDDINGS.get(nueva_pregunta.lower())
    
    if not nuevo_vector:
        return None, 0.0 # Pregunta totalmente nueva que no mapeamos en la simulación
        
    mejor_similitud = 0.0
    mejor_coincidencia = None
    
    # 2. Escanear la base de datos vectorial buscando la pregunta más cercana
    for pregunta_guardada, datos in BASE_DATOS_VECTORIAL_CACHE.items():
        similitud = calcular_similitud_coseno(nuevo_vector, datos["vector"])
        if similitud > mejor_similitud:
            mejor_similitud = similitud
            mejor_coincidencia = datos["respuesta_llm"]
            
    # 3. Decidir si la similitud pasa el umbral (ej. 95% de coincidencia semántica)
    if mejor_similitud >= umbral_tolerancia:
        return mejor_coincidencia, mejor_similitud
        
    return None, mejor_similitud

def procesar_peticion_usuario(pregunta_usuario: str):
    print(f"📥 Entrada del usuario: '{pregunta_usuario}'")
    inicio = time.time()
    
    # Intentar resolver mediante Caché Semántico primero
    respuesta_cache, score = buscar_en_cache_semantico(pregunta_usuario)
    
    if respuesta_cache:
        print(f"🎯 ¡HIT DE CACHÉ SEMÁNTICO! (Similitud del {score*100:.2f}%)")
        print(f"⚡ Tiempo de respuesta: {time.time() - inicio:.4f}s")
        print(f"💰 Costo de Tokens: $0.000000 USD (Reutilizado localmente)")
        print(f"🤖 Respuesta: {respuesta_cache}\n")
    else:
        print(f"❌ MISS DE CACHÉ. (Similitud máxima fue de apenas {score*100:.2f}%)")
        print("⏳ Conectando con el LLM en la nube para generar una nueva respuesta...")
        time.sleep(1.8) # Simula la latencia y el procesamiento del LLM
        
        # Simulación de respuesta generada por el LLM externo
        nueva_respuesta = "El precio de nuestro plan mensual es de $29 USD e incluye acceso ilimitado a todos los módulos."
        
        # GUARDAR EN EL CACHÉ VECTORIAL para la siguiente ocasión
        vector_nuevo = DICCIONARIO_EMBEDDINGS.get(pregunta_usuario.lower(), [0.0, 0.0, 0.0])
        BASE_DATOS_VECTORIAL_CACHE[pregunta_usuario.lower()] = {
            "vector": vector_nuevo,
            "respuesta_llm": nueva_respuesta
        }
        
        print(f"⚡ Tiempo de respuesta: {time.time() - inicio:.4f}s")
        print(f"💸 Costo de Tokens: ~$0.004500 USD (Llamada a API Cloud efectuada)")
        print(f"🤖 Respuesta: {nueva_respuesta}\n")

# --- FLUJO DE PRUEBA ---
if __name__ == "__main__":
    print("🖥️ Inicializando Sistema con Optimización Vectorial...\n")
    
    # Petición 1: El usuario pregunta algo semánticamente idéntico a lo que ya está en caché, pero con diferentes palabras.
    procesar_peticion_usuario("¿De qué forma puedo cambiar la contraseña?")
    
    # Petición 2: El usuario pregunta algo que nunca nadie ha consultado antes.
    procesar_peticion_usuario("¿Cuánto cuesta la suscripción mensual?")
    
    # Petición 3: El usuario vuelve a preguntar lo mismo de la Petición 2, pero usando sinónimos.
    procesar_peticion_usuario("¿Cuál es el precio del plan por mes?")
```

## 1. El Corazón de la IA en AWS: Amazon Bedrock

**Amazon Bedrock** es el servicio estrella de AWS para Inteligencia Artificial Generativa. En lugar de tener que lidiar con las APIs individuales de OpenAI, Anthropic, Cohere o Meta, AWS te da una única interfaz unificada para consumirlos todos.

### Opciones de Compra y Ahorro en Bedrock:

* **On-Demand (Bajo Demanda):** Pagas estrictamente por token procesado (lo que calculamos en la Fase 1). Es perfecto para prototipos y tráfico variable.
* **Batch (Procesamiento por lotes):** Envías un archivo con miles de prompts por la noche. AWS los procesa cuando sus servidores tienen menos carga y te cobra hasta un **50% o 70% menos** por token. Ideal para análisis de datos masivos que no requieren respuesta inmediata.
* **Provisioned Throughput (Rendimiento aprovisionado):** Reservas unidades de cómputo dedicadas por un tiempo fijo. Solo sirve si tienes un tráfico masivo constante, de lo contrario, desperdicias dinero.

---

## 2. Los Componentes del Ahorro en Producción

Para implementar lo que programamos en las Fases 2 y 3 dentro de AWS, utilizamos los siguientes servicios:

1. **AWS Lambda (Serverless Compute):** Aquí vive nuestro código Python de enrutamiento asíncrono (`asyncio`). Como Lambda solo cobra por los milisegundos que está corriendo, si el caché semántico responde en 0.05 segundos, la infraestructura te cuesta prácticamente cero.
2. **Amazon ElastiCache (Redis OSS) con OpenSearch Serverless:** Son los servicios ideales para almacenar nuestra base de datos vectorial de "bajo impacto". Tienen capacidades nativas para calcular la *Similitud de Coseno* a velocidad luz (microsegundos).
3. **Amazon DynamoDB:** Una base de datos NoSQL ultra rápida donde guardamos el historial de las conversaciones de los usuarios de forma económica.

````python
import json
import boto3

# Inicializamos el cliente oficial de AWS para interactuar con los modelos de IA
# (En producción, AWS Lambda maneja esto de forma interna y segura)
try:
    bedrock_client = boto3.client(service_name="bedrock-runtime", region_name="us-east-1")
except Exception:
    bedrock_client = None # Modo simulación si corres local sin llaves de AWS

def evaluar_complejidad_local(prompt: str) -> str:
    """Heurística rápida de bajo costo antes de llamar a Bedrock"""
    if len(prompt.split()) > 15 or "analiza" in prompt.lower() or "código" in prompt.lower():
        return "COMPLEJA"
    return "SIMPLE"

def invocar_modelo_bedrock(prompt: str, tier: str):
    """
    Simula o ejecuta la llamada real a Amazon Bedrock usando 
    el identificador único del modelo (Model ID).
    """
    # Definimos los IDs de los modelos en AWS Bedrock
    ID_MODELO_BARATO = "anthropic.claude-3-haiku-20240307-v1:0"
    ID_MODELO_AVANZADO = "anthropic.claude-3-5-sonnet-20240620-v1:0"
    
    model_id = ID_MODELO_BARATO if tier == "SIMPLE" else ID_MODELO_AVANZADO
    
    print(f"📡 Enviando solicitud a AWS Bedrock -> Modelo: {model_id}")
    
    # Si tienes AWS configurado localmente, esto intentará hacer la llamada real
    if bedrock_client:
        try:
            body = json.dumps({
                "anthropic_version": "bedrock-2023-05-31",
                "max_tokens": 500,
                "messages": [{"role": "user", "content": prompt}]
            })
            response = bedrock_client.invoke_model(body=body, modelId=model_id)
            response_body = json.loads(response.get('body').read())
            return response_body['content'][0]['text']
        except Exception as e:
            return f"Error de conexión real a AWS: {str(e)}"
    else:
        # Caída elegante (Fallback) para nuestra simulación local
        return f"[Simulación AWS] Respuesta exitosa generada por Bedrock usando {model_id.split('.')[1]}"

def lambda_handler(event, context):
    """
    Esta es la función principal que AWS ejecuta automáticamente 
    cada vez que un usuario hace una petición desde la app.
    """
    # 1. Extraer el mensaje que envió el usuario
    body = json.loads(event.get("body", "{}"))
    prompt_usuario = body.get("prompt", "¿Hola, cómo estás?")
    
    # 2. Paso 1: (Simulado de Fase 3) -> ¿Está en el caché vectorial?
    # En producción aquí harías un query rápido a tu Redis/OpenSearch local.
    cache_hit = False 
    
    if cache_hit:
        return {
            "statusCode": 200,
            "body": json.dumps({"origen": "Caché Local", "respuesta": "Respuesta guardada"})
        }
        
    # 3. Paso 2: Arbitraje de Modelos (Fase 2)
    complejidad = evaluar_complejidad_local(prompt_usuario)
    
    # 4. Paso 3: Consumir la IA en AWS de manera eficiente
    respuesta_ia = invocar_modelo_bedrock(prompt_usuario, complejidad)
    
    return {
        "statusCode": 200,
        "body": json.dumps({
            "origen": f"Amazon Bedrock ({complejidad})",
            "respuesta": respuesta_ia
        }, ensure_ascii=False)
    }

# --- EJECUCIÓN LOCAL DE PRUEBA ---
if __name__ == "__main__":
    print("☁️ Probando lógica de AWS Lambda localmente...\n")
    
    # Simulación de evento de entrada de API Gateway (Petición simple)
    evento_simple = {"body": json.dumps({"prompt": "Buenas tardes"})}
    resultado_1 = lambda_handler(evento_simple, None)
    print(f"Resultado: {resultado_1['body']}\n")
    
    # Simulación de petición compleja
    evento_complejo = {"body": json.dumps({"prompt": "Analiza las tendencias de este bloque de código y optimiza el performance"})}
    resultado_2 = lambda_handler(evento_complejo, None)
    print(f"Resultado: {resultado_2['body']}\n")
````
