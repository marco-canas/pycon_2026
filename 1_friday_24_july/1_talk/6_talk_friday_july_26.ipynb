###  Detección y Limpieza de Anomalías (Código Práctico)

A continuación tienes un script completo que simula lecturas de un reactor termoquímico con errores reales (valores faltantes por fallo de sensor y lecturas erróneas por picos de tensión) y muestra cómo corregirlos.



````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. GENERACIÓN DE DATOS SINTÉTICOS CON ERRORES
np.random.seed(42)
n_samples = 100

data = {
    'Tiempo_min': np.arange(1, n_samples + 1),
    'Temperatura_K': np.random.normal(loc=850, scale=15, size=n_samples), # Temp alrededor de 850 K
    'Presion_bar': np.random.normal(loc=25, scale=2, size=n_samples),       # Presión alrededor de 25 bar
    'Flujo_Metano_mol_h': np.random.normal(loc=10, scale=0.5, size=n_samples)
}

df = pd.DataFrame(data)

# Introducimos problemas típicos de laboratorio:
# a) Valores faltantes (NaN) por pérdida de señal
df.loc[15:18, 'Temperatura_K'] = np.nan

# b) Outliers (picos erróneos por fallo de sensor)
df.loc[50, 'Temperatura_K'] = 1500  # Imposiblemente alta
df.loc[75, 'Presion_bar'] = -5       # Imposiblemente baja

print("--- Primeras 5 filas con datos crudos ---")
print(df.head())

# 2. TRATAMIENTO DE ERRORES

# a) Imputación de valores faltantes mediante interpolación lineal (ideal para series de tiempo)
df['Temperatura_K'] = df['Temperatura_K'].interpolate(method='linear')

# b) Filtrado de Outliers aplicando límites físicos/operativos conocidos
temp_max_fisica = 1100 # Kelvin
presion_min_fisica = 0  # Bar

# Reemplazamos los valores fuera de rango por NaN y luego interpolamos o descartamos
df.loc[df['Temperatura_K'] > temp_max_fisica, 'Temperatura_K'] = np.nan
df.loc[df['Presion_bar'] < presion_min_fisica, 'Presion_bar'] = np.nan

df['Temperatura_K'] = df['Temperatura_K'].ffill() # Forward fill para completar
df['Presion_bar'] = df['Presion_bar'].ffill()

# 3. CREACIÓN DE VARIABLES FISICOQUÍMICAS DERIVADAS
# Ejemplo: Calculemos la Densidad molar aproximada usando la ley de gases ideales (P*V = n*R*T -> n/V = P / (R*T))
R = 0.08314  # bar * L / (mol * K)
df['Densidad_Molar_mol_L'] = df['Presion_bar'] / (R * df['Temperatura_K'])

print("\n--- Resumen estadístico tras la limpieza ---")
print(df[['Temperatura_K', 'Presion_bar', 'Densidad_Molar_mol_L']].describe())
````

# Solución del ejercicio anterior  

Aquí tienes el script completo en Python para resolver el reto práctico.

Incluye la generación del conjunto de datos simulado (para que el código sea 100% ejecutable por sí solo), la corrección de los valores cero, el cálculo de la variable fisicoquímica y la visualización final.


````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de semilla para reproducibilidad
np.random.seed(42)
n_muestras = 100

# -----------------------------------------------------------------------------
# 0. GENERACIÓN DEL DATASET SIMULADO (Para simular el archivo CSV de entrada)
# -----------------------------------------------------------------------------
datos_gasificador = {
    'Temperatura_K': np.random.normal(loc=1050, scale=30, size=n_muestras),
    'CO_pct_mol': np.random.normal(loc=20, scale=3, size=n_muestras),
    'CO2_pct_mol': np.random.normal(loc=15, scale=2, size=n_muestras),
    'H2_pct_mol': np.random.normal(loc=35, scale=4, size=n_muestras),
    'CH4_pct_mol': np.random.normal(loc=8, scale=1, size=n_muestras)
}

df_syngas = pd.DataFrame(datos_gasificador)

# Introducimos intencionalmente ceros en CO para simular errores de lectura del sensor
df_syngas.loc[[10, 45, 82], 'CO_pct_mol'] = 0.0

print("--- Primeras 5 filas del dataset original ---")
print(df_syngas.head())

# -----------------------------------------------------------------------------
# PASO 1 Y 2: Manejo de división por cero y cálculo de la relación H2/CO
# -----------------------------------------------------------------------------

# Identificamos cuántos ceros hay en la columna CO
ceros_co = (df_syngas['CO_pct_mol'] == 0).sum()
print(f"\n[Aviso] Se detectaron {ceros_co} valores iguales a 0 en la columna 'CO_pct_mol'.")

# Calculamos la mediana excluyendo los valores iguales a cero
mediana_co = df_syngas[df_syngas['CO_pct_mol'] > 0]['CO_pct_mol'].median()

# Sustituimos los ceros por la mediana para prevenir la división por cero (Inf/NaN)
df_syngas['CO_pct_mol'] = df_syngas['CO_pct_mol'].replace(0.0, mediana_co)

# Calculamos la relación molar H2 / CO
df_syngas['Relacion_H2_CO'] = df_syngas['H2_pct_mol'] / df_syngas['CO_pct_mol']

print("\n--- Resumen estadístico tras la corrección y cálculo de H2/CO ---")
print(df_syngas[['Temperatura_K', 'CO_pct_mol', 'H2_pct_mol', 
                 'Relacion_H2_CO']].describe())

# -----------------------------------------------------------------------------
# PASO 3: Generación del gráfico de dispersión (Scatter Plot)
# -----------------------------------------------------------------------------
sns.set_theme(style="whitegrid")
plt.figure(figsize=(9, 5))

# Creación del Scatter Plot con seaborn
sns.scatterplot(
    data=df_syngas,
    x='Temperatura_K',
    y='Relacion_H2_CO',
    hue='CH4_pct_mol',  # Color basado en CH4 para añadir una 3ra dimensión útil
    palette='viridis',
    s=70,               # Tamaño de los puntos
    alpha=0.8
)

# Personalización del gráfico
plt.title('Relación Molar $H_2/CO$ vs. Temperatura del Gasificador', 
          fontsize=13, pad=12, weight='bold')
plt.xlabel('Temperatura ($K$)', fontsize=11)
plt.ylabel('Relación Molar $H_2 / CO$', fontsize=11)

# Añadimos una línea de referencia para calidad estándar de syngas (ej. H2/CO = 2)
plt.axhline(y=2.0, color='red', linestyle='--', linewidth=1.2, 
            label='Relación Teórica Objetivo ($H_2/CO = 2.0$)')

plt.legend(title='$\%mol\ CH_4$', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig(r"C:\Users\marco\Downloads\relacion_H2_CO_vs_Temperatura.png", dpi=300)
plt.show()
````


# **Detalles clave de la solución:**

1. **Evitar la contaminación del cálculo:** Para calcular la `mediana_co`, primero filtramos los ceros (`df_syngas['CO_pct_mol'] > 0`). Si incluyes los ceros en el cálculo de la mediana, podrías sesgar levemente el valor hacia abajo.
2. **Sustitución explícita:** Se usa el método `.replace(0.0, mediana_co)` de pandas, el cual reemplaza de forma eficiente los valores nulos o atípicos de la serie.
3. **Visualización enriquecida:** Además de graficar la temperatura frente a la relación $H_2/CO$, agregamos el parámetro `hue='CH4_pct_mol'`, lo que te permite ver en la misma gráfica de 2D una tercera variable química usando un mapa de colores.

### **1. Conceptos Térmicos y Reacciones de Gasificación/Reformado**

En procesos termoquímicos de producción de hidrógeno y *syngas* (gas de síntesis), las reacciones clave son:

1. **Reacción de Reformado con Vapor (*Steam Methane Reforming - SMR*):**

$$\text{CH}_4 + \text{H}_2\text{O} \rightleftharpoons \text{CO} + 3\text{H}_2 \quad (\Delta H^\circ > 0, \text{endotérmica})$$


2. **Reacción de Desplazamiento del Gas de Agua (*Water-Gas Shift - WGS*):**

$$\text{CO} + \text{H}_2\text{O} \rightleftharpoons \text{CO}_2 + \text{H}_2 \quad (\Delta H^\circ < 0, \text{exotérmica})$$



**Parámetros clave a calcular:**

* **Conversión de metano ($X_{\text{CH}_4}$):** $X_{\text{CH}_4} = \frac{n_{\text{CH}_4, \text{in}} - n_{\text{CH}_4, \text{out}}}{n_{\text{CH}_4, \text{in}}}$
* **Rendimiento de Hidrógeno ($Y_{\text{H}_2}$):** $Y_{\text{H}_2} = \frac{n_{\text{H}_2, \text{out}}}{n_{\text{CH}_4, \text{in}}}$

### **2. Simulación Cinética de un Reactor Continuo (Código Práctico)**

A continuación, implementaremos un modelo cinético simplificado de un **Reactor PBR (Packed Bed Reactor)** donde ocurre la reacción de reformado de metano. Resolveremos la ecuación diferencial del balance de masa en el reactor usando `scipy.integrate.solve_ivp`.


````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# -----------------------------------------------------------------------------
# 1. DEFINICIÓN DE CONSTANTES Y MODELO CINÉTICO
# -----------------------------------------------------------------------------
# Parámetros cinéticos (Ecuación de Arrhenius: k = A * exp(-Ea / (R * T)))
R_gas = 8.314 J / (mol * K)
A = 1.2e5       # Factor preexponencial (m^3 / (kg_cat * h))
Ea = 65000      # Energía de activación (J/mol)

def k_reaccion(T_K):
    """Calcula la constante de velocidad k a una Temperatura dada."""
    return A * np.exp(-Ea / (R_gas * T_K))

def balances_materia(w, Y, T_K):
    """
    Ecuaciones diferenciales del balance de masa en función de la masa de catalizador (w).
    Y = [F_CH4, F_H2O, F_CO, F_H2] (Flujos molares en mol/h)
    """
    F_CH4, F_H2O, F_CO, F_H2 = Y
    
    # Concentración/Presión parcial simplificada (asumiendo volumen constante)
    F_total = F_CH4 + F_H2O + F_CO + F_H2
    C_CH4 = F_CH4 / F_total
    C_H2O = F_H2O / F_total
    
    # Velocidad de reacción r_SMR = k * C_CH4 * C_H2O
    r = k_reaccion(T_K) * C_CH4 * C_H2O
    
    # Estequiometría: CH4 + H2O -> CO + 3 H2
    dF_CH4_dw = -r
    dF_H2O_dw = -r
    dF_CO_dw  = +r
    dF_H2_dw  = +3 * r
    
    return [dF_CH4_dw, dF_H2O_dw, dF_CO_dw, dF_H2_dw]

# -----------------------------------------------------------------------------
# 2. CONFIGURACIÓN Y RESOLUCIÓN DE LA SIMULACIÓN
# -----------------------------------------------------------------------------
# Condiciones de entrada al reactor
F_CH4_0 = 10.0  # mol/h
F_H2O_0 = 30.0  # mol/h (Relación Vapor/Carbono S/C = 3)
F_CO_0  = 0.0
F_H2_0  = 0.0
Y0 = [F_CH4_0, F_H2O_0, F_CO_0, F_H2_0]

T_operacion = 950.0  # Kelvin (Isotérmico)
W_catalizador = np.linspace(0, 50, 100) # De 0 a 50 kg de catalizador

# Resolver el sistema de ODEs
solucion = solve_ivp(
    fun=lambda w, Y: balances_materia(w, Y, T_operacion),
    t_span=(W_catalizador[0], W_catalizador[-1]),
    y0=Y0,
    t_eval=W_catalizador,
    method='RK45'
)

# Convertir la solución a un DataFrame de Pandas
df_simulacion = pd.DataFrame({
    'Masa_Catalizador_kg': solucion.t,
    'F_CH4': solucion.y[0],
    'F_H2O': solucion.y[1],
    'F_CO': solucion.y[2],
    'F_H2': solucion.y[3]
})

# Calcular Conversión de CH4 y Rendimiento de H2
df_simulacion['Conversion_CH4'] = (F_CH4_0 - df_simulacion['F_CH4']) / F_CH4_0
df_simulacion['Rendimiento_H2'] = df_simulacion['F_H2'] / F_CH4_0

print("--- Perfil final a la salida del reactor (50 kg catalizador) ---")
print(df_simulacion.iloc[-1][['Masa_Catalizador_kg', 'Conversion_CH4', 'Rendimiento_H2']])
````

# Solución del reto práctico de la semana 2  

Aquí tienes el script completo en Python para resolver el reto práctico de la Semana 2.

En este script iteramos sobre un rango de temperaturas, ejecutamos la simulación cinética en cada iteración y almacenamos los resultados clave. Al final, visualizamos la sensibilidad térmica y exportamos los datos a un archivo CSV que servirá como el *dataset* perfecto para entrenar los modelos de Machine Learning en la **Semana 3**.


````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.integrate import solve_ivp

# -----------------------------------------------------------------------------
# 1. DEFINICIÓN DEL MODELO CINÉTICO Y FUNCIONES
# -----------------------------------------------------------------------------
R_gas = 8.314  # J / (mol * K)
A = 1.2e5      # Factor preexponencial (m^3 / (kg_cat * h))
Ea = 65000     # Energía de activación (J/mol)

def k_reaccion(T_K):
    """Calcula la constante de velocidad k a una Temperatura dada."""
    return A * np.exp(-Ea / (R_gas * T_K))

def balances_materia(w, Y, T_K):
    """
    Ecuaciones diferenciales del balance de masa en el reactor.
    Y = [F_CH4, F_H2O, F_CO, F_H2]
    """
    F_CH4, F_H2O, F_CO, F_H2 = Y
    F_total = F_CH4 + F_H2O + F_CO + F_H2
    
    # Fracciones molares
    C_CH4 = F_CH4 / F_total
    C_H2O = F_H2O / F_total
    
    # Velocidad de reacción r_SMR
    r = k_reaccion(T_K) * C_CH4 * C_H2O
    
    # Balances por componente
    dF_CH4_dw = -r
    dF_H2O_dw = -r
    dF_CO_dw  = +r
    dF_H2_dw  = +3 * r
    
    return [dF_CH4_dw, dF_H2O_dw, dF_CO_dw, dF_H2_dw]

# -----------------------------------------------------------------------------
# 2. BUCLE DE SIMULACIÓN A DIFERENTES TEMPERATURAS
# -----------------------------------------------------------------------------
# Condiciones iniciales y del lecho catalítico
F_CH4_0 = 10.0  # mol/h
F_H2O_0 = 30.0  # mol/h (Relación S/C = 3)
Y0 = [F_CH4_0, F_H2O_0, 0.0, 0.0]
W_catalizador = (0, 50)  # De 0 a 50 kg de catalizador

# Evaluaremos un rango amplio de temperaturas para generar datos variados (700 K a 1100 K)
temperaturas_evaluar = np.linspace(700, 1100, 40)
resultados_sensibilidad = []

for T in temperaturas_evaluar:
    # Resolver la ED para la temperatura actual T
    sol = solve_ivp(
        fun=lambda w, Y: balances_materia(w, Y, T),
        t_span=W_catalizador,
        y0=Y0,
        method='RK45'
    )
    
    # Extraer valores finales a la salida del reactor (W = 50 kg)
    F_CH4_salida = sol.y[0][-1]
    F_H2_salida  = sol.y[3][-1]
    
    # Calcular métricas de desempeño
    conversion_CH4 = (F_CH4_0 - F_CH4_salida) / F_CH4_0
    rendimiento_H2 = F_H2_salida / F_CH4_0
    
    # Guardar en la lista
    resultados_sensibilidad.append({
        'Temperatura_K': T,
        'F_CH4_salida_mol_h': F_CH4_salida,
        'F_H2_salida_mol_h': F_H2_salida,
        'Conversion_CH4': conversion_CH4,
        'Rendimiento_H2': rendimiento_H2
    })

# Convertir a DataFrame
df_resultados = pd.DataFrame(resultados_sensibilidad)

print("--- Primeras y últimas filas del conjunto de datos generado ---")
print(df_resultados[['Temperatura_K', 'Conversion_CH4', 'Rendimiento_H2']].head())
print(df_resultados[['Temperatura_K', 'Conversion_CH4', 'Rendimiento_H2']].tail())

# Guardar el dataset generado a un archivo CSV para la Semana 3
df_resultados.to_csv('datos_simulacion_termoquimica.csv', index=False)
print("\n[OK] Dataset guardado exitosamente como 'datos_simulacion_termoquimica.csv'")

# -----------------------------------------------------------------------------
# 3. VISUALIZACIÓN DE LA SENSIBILIDAD TÉRMICA
# -----------------------------------------------------------------------------
sns.set_theme(style="whitegrid")
fig, ax1 = plt.subplots(figsize=(9, 5))

color_temp = 'tab:red'
ax1.set_xlabel('Temperatura de Operación (K)', fontsize=11)
ax1.set_ylabel('Conversión de $CH_4$', color=color_temp, fontsize=11)
ax1.plot(df_resultados['Temperatura_K'], df_resultados['Conversion_CH4'], color=color_temp, lw=2.5, label='Conversión $CH_4$')
ax1.tick_params(axis='y', labelcolor=color_temp)

# Eje secundario para el Rendimiento de H2
ax2 = ax1.twinx()
color_h2 = 'tab:blue'
ax2.set_ylabel('Rendimiento de $H_2$ (mol $H_2$ / mol $CH_4$)', color=color_h2, fontsize=11)
ax2.plot(df_resultados['Temperatura_K'], df_resultados['Rendimiento_H2'], color=color_h2, lw=2.5, linestyle='--', label='Rendimiento $H_2$')
ax2.tick_params(axis='y', labelcolor=color_h2)

plt.title('Sensibilidad Térmica: Conversión y Rendimiento vs. Temperatura', fontsize=13, weight='bold', pad=12)
fig.tight_layout()
plt.savefig(r"C:\Users\marco\Downloads\sensibilidad_termica_conversion_rendimiento.png", dpi=300)
plt.show()
````
