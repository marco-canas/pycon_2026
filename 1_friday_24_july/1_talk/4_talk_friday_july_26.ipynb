## Resumen de la charla

¿Qué pasaría si el navegador pudiera ejecutar Python como un lenguaje de primera clase? 

En esta charla, mostraré cómo PyScript hace posible ejecutar Python real directamente en el navegador, gracias al poder de WebAssembly.



A través de una serie de emocionantes ejemplos en vivo, verás cómo Python manipula el DOM, llama a las API del navegador y crea experiencias interactivas, todo sin necesidad de una base de código tradicional en JavaScript. 

También mostraré un par de ejemplos de cómo puedes integrar tanto JavaScript como Python en PyScript para crear herramientas aún más interesantes.



Además, analizaré qué es WebAssembly, por qué existe y cómo permite que lenguajes como Python se ejecuten de forma segura y eficiente en la plataforma web. 

Por último, abordaré cuándo tiene sentido utilizar herramientas como PyScript y la compararé con opciones similares.



Ya seas un desarrollador de Python con curiosidad por el *frontend*, un ingeniero interesado en WebAssembly o simplemente alguien que disfruta ver cómo se desafían los límites de Python, esta charla cambiará tu forma de pensar sobre lo que se puede ejecutar en un navegador.

## Programa del Curso: "Python en la Web con WebAssembly y PyScript"

1. **Módulo 1: Fundamentos de WebAssembly (WASM) y Pyodide:** 3 - 4 horas.
**Objetivo:** Entender la tecnología subyacente que permite a lenguajes no web correr en el navegador.

* **Conceptos clave:**
* ¿Qué es WebAssembly (WASM)? Formato de instrucción binario y sandbox de seguridad.
* Entender la diferencia entre interpretar JavaScript vs. ejecutar un binario WASM.
* **Pyodide:** La distribución de CPython portada a WebAssembly con Emscripten.




* **Ejercicio práctico:** Crear un documento HTML simple e importar Pyodide mediante un script `<script>` de JavaScript para ejecutar un `print("Hola desde WASM")` básico en la consola del navegador.




2. **Módulo 2: Primeros pasos con PyScript:** 4 - 5 horas.
**Objetivo:** Dominar las etiquetas y configuración de PyScript para escribir Python directo en HTML.

* **Conceptos clave:**
* Etiquetas fundamentales: `<script type="py">` y configuración mediante `pyscript.json` (o `<py-config>`).
* Manejo de dependencias: Cómo instalar paquetes del ecosistema de Python (`numpy`, `pandas`, etc.) dentro del navegador.
* Ciclo de vida de la aplicación e inicialización del entorno runtime.


* **Ejercicio práctico:** Crear una app interactiva que cargue una librería como `math` o `json` y muestre resultados en pantalla al renderizar el HTML.

**Objetivo:** Manipular la interfaz gráfica y comunicar JavaScript con Python sin problemas de contexto.

* **Conceptos clave:**
* Módulo `pyscript.web` y la interfaz `pyodide.ffi`: cómo acceder a `document`, `window` y llamadas Fetch.
* Manejo de eventos: Vincular botones (`onClick`) o inputs a funciones escritas en Python.
* Interoperabilidad bidireccional: Pasar datos de JavaScript a Python y viceversa.


* **Ejercicio práctico:** Construir un dashboard simple o una calculadora interactiva donde los eventos de los botones activen funciones en Python que modifiquen el texto o estilo de la página.


4. **Módulo 4: Casos de uso, Pros, Contras y Alternativas:** 3 horas.
**Objetivo:** Desarrollar el criterio técnico para saber *cuándo usar* estas tecnologías y cuándo no.

* **Conceptos clave:**
* **Rendimiento y costo:** Tamaño de descarga inicial (Carga del motor de Python ~10MB+) vs. tiempo de ejecución.
* **Comparativa:** PyScript vs. Anvil vs. Brython vs. aplicaciones SPA tradicionales con React/Vue.
* Casos de uso ideales: Dashboards científicos, demos de Machine Learning offline, herramientas educativas y prototipado rápido.
