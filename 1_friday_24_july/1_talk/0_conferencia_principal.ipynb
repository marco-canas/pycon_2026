# Módulo 1: La Evolución del Manejo de Excepciones

En Python tradicional, una función o bloque de código solo podía lanzar o capturar **una excepción a la vez**. Sin embargo, con el auge de la programación asíncrona y la ejecución en paralelo (múltiples tareas ejecutándose al mismo tiempo), es muy común que **varias cosas fallen al mismo tiempo**.

---



## 1. El Problema: Excepciones en Concurrencia

Imagina que lanzas tres peticiones HTTP en paralelo con `asyncio`. Si dos de ellas fallan por distintas razones (una por tiempo de espera y otra por error 404), la forma antigua de Python te obligaba a elegir cuál excepción lanzar o envolverlas en clases personalizadas poco estandarizadas.

```
                  ┌─► Tarea 1: OK
                  │
[TaskGroup] ──────┼─► Tarea 2: ConnectionError  ──┐
                  │                               ├─► ¿Cómo capturamos AMBOS?
                  └─► Tarea 3: FileNotFoundError ──┘

```

---



## 2. La Solución: `ExceptionGroup`

Un `ExceptionGroup` es una excepción que **empaqueta varias excepciones dentro de sí misma** de forma jerárquica (puede incluir dentro otros grupos de excepciones).

### Creación manual de un grupo


````python
# Crear un grupo de excepciones manualmente
errores = ExceptionGroup(
    "Fallaron varios procesos de red",
    [
        ConnectionResetError("El servidor cerró la conexión"),
        TimeoutError("El servidor no respondió a tiempo"),
    ]
)

# Puedes inspeccionar sus excepciones hijas
print(errores.message)    # "Fallaron varios procesos de red"
print(errores.exceptions) # (ConnectionResetError(...), TimeoutError(...))
````


# 3. Captura Selectiva con `except*`

Para no romper el comportamiento de los bloques `try...except` tradicionales, Python 3.11 introdujo el operador `except*`.

A diferencia del `except` tradicional (que se ejecuta solo una vez y detiene la búsqueda de otras cláusulas), **varios bloques `except*` se pueden ejecutar para el mismo grupo de excepciones**, "desglosando" el paquete según el tipo de error.



````python
try:
    raise ExceptionGroup(
        "Errores en la base de datos y en disco",
        [
            ValueError("Dato inválido"),
            FileNotFoundError("Archivo de configuración no encontrado"),
            TypeError("Tipo de dato no soportado")
        ]
    )

except* ValueError as e:
    print(f"Manejando los errores de valor: {e.exceptions}")

except* FileNotFoundError as e:
    print(f"Manejando archivos no encontrados: {e.exceptions}")

except* TypeError as e:
    print(f"Manejando errores de tipo: {e.exceptions}")
````


> **¿Qué ocurre internamente?**
> El motor de Python divide el `ExceptionGroup` original. Si un bloque `except*` no captura todos los errores del grupo, los no capturados se vuelven a empaquetar automáticamente y continúan su viaje hacia arriba en la pila de llamadas.

---

## 4. Uso Práctico con `asyncio.TaskGroup`

El caso de uso estrella de esta funcionalidad es `TaskGroup`, la forma moderna de ejecutar tareas asíncronas en Python.



````python
import asyncio # este paquete sirve para manejar tareas asíncronas, es decir, tareas que se ejecutan en paralelo y no bloquean el flujo del programa

async def tarea_fallida_1():
    await asyncio.sleep(0.1)
    raise ConnectionError("Fallo en la red")

async def tarea_fallida_2():
    await asyncio.sleep(0.2)
    raise ValueError("Parámetro inválido")

async def main():
    try:
        async with asyncio.TaskGroup() as tg:
            tg.create_task(tarea_fallida_1())
            tg.create_task(tarea_fallida_2())

    except* ConnectionError:
        print("Reintentando conexión...")

    except* ValueError:
        print("Revisando los parámetros de entrada...")

await main()
````


## 5. Rendimiento Interno: *Zero-Cost Exceptions*

Un detalle clave del trabajo del equipo de CPython (del cual Irit forma parte) es cómo hicieron que los bloques `try` fueran **gratuitos en rendimiento** cuando no ocurren errores:

* **Antes:** Entrar a un bloque `try` creaba un marco o registro en la memoria de la pila de Python por si acaso ocurría una excepción.
* **Ahora (*Zero-Cost*):** El compilador genera una **tabla de excepciones interna** (*Exception Table*) en el bytecode. El código se ejecuta a máxima velocidad dentro del `try` sin costo adicional. Solo si se detecta un error, el intérprete consulta esa tabla estática para saber a qué instrucción saltar.
